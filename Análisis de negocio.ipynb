```python
# Crear la API
from fastapi import FastAPI
from pydantic import BaseModel
import pandas as pd
import joblib
```

```python
# Cargar el modelo
modelo = joblib.load('modelo_retencion_pipeline.pkl')
```

```python
# Definir app
app = FastAPI()
```

```python
# Datos que pide la API
class Cliente(BaseModel):
    producto_comercial: str
    dgt_agrupado: str
    forma_contratacion: str
    tipo_productor: str
    forma_pago: str
    cliente_activo_digital: str
    tramo_edad: str
    tramo_antiguedad: str
    canal_contratacion_agrupado: str
    integralidad_cliente: str
    canal_cliente: float
    sectores_estrategicos: float
    tipo_valor_cliente: int
    mes_proceso: int
    tramo_numero_polizas: int
    sectores_contratados: str # texto, luego lo convertimos a longitud
```

```python
@app.post("/predecir")
def predecir(cliente: Cliente):
    entrada = cliente.dict()
    # Calcular longitud de sectores_contratados
    entrada['sectores_contratados_codificado'] = len(entrada['sectores_contratados'])
    # Eliminar campo original
    del entrada['sectores_contratados']
    # Convertir a DataFrame
    datos = pd.DataFrame([entrada])
    # Predecir con el pipeline
    probabilidad = modelo.predict_proba(datos)[0][1]
    return {"probabilidad_de_ser_retenido": round(probabilidad, 4),
            "clasificacion": "RETENER" if probabilidad >= 0.51 else "RIESGO ABANDONO"}
```

```python
uvicorn.run(app, host="127.0.0.1", port=8000)
```
